## Problem 2 - Multiple value records in country column

1. Some values in country column has multiple countries.
2. Lets get no:of records for each country.
3. Using a dictionary can solve the problem. 
4. The keys have spaces making them as 2 keys for same country. 
    Example :
    - 'Israel': 22,
    - ' Israel': 8
    - Solution - Adding strip() in dict_func()

```python
def dict_func(val):
    val = val.strip() #4 Solution
    if val not in country_values.keys():
        country_values[val]=1
    else:
        country_values[val]+=1
```

```python
country_value = df['country']
country_values ={}
for i in country_value:
    if "," in i:
        sep = i.split(",")
        for j in sep:
            dict_func(j)
    else:
        dict_func(i)
        
country_values
```

## Problem 3 - Multiple Genre records in column listed_in

1. Some values in this column has multiple genres.
2. Lets get no:of records for each genre.
3. Using a dictionary can solve the problem. 

```python
def dict_func2(val):
    val = val.strip() #4 Solution
    if val not in genre_values.keys():
        genre_values[val]=1
    else:
        genre_values[val]+=1
```

```python
genre_value = df['listed_in']
genre_value
genre_values ={}
for i in genre_value:
    if "," in i:
        sep = i.split(",")
        for j in sep:
            dict_func2(j)
    else:
        dict_func2(i)
        
genre_values
```

## Viz 3 - year wise count of shows or movies in line graph

```python
df_movies = df[df['type']=='Movie']
df_tvshows = df[df['type']=='TV Show']
year_movies = df_movies.groupby('release_year')['type'].count()
year_tvshows = df_tvshows.groupby('release_year')['type'].count()

fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(year_movies.index, year_movies.values, label='Movies', color='#e50914', linewidth=2.5)
ax.plot(year_tvshows.index, year_tvshows.values, label='TV Shows', color='#221f1f', linewidth=2.5)

ax.set_title('Number of Movies and TV Shows Released Each Year', fontsize=16, fontweight='bold', fontfamily='serif', color='#221f1f')
ax.set_xlabel('Release Year', fontsize=14, fontfamily='serif', color='#221f1f')
ax.set_ylabel('Number of Releases', fontsize=14, fontfamily='serif', color='#221f1f')
ax.tick_params(axis='x', labelsize=12, colors='#221f1f')
ax.tick_params(axis='y', labelsize=12, colors='#221f1f')

ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.7)

for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)
ax.legend(loc='upper left', fontsize=12)
plt.show()
```
